```python
from sklearn import datasets
import pandas as pd
df = pd.read_csv('lottery.csv')
feature = df[['first','second', 'third']]
feature
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = KMeans(n_clusters=4, algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['predict']
```

```python
r = pd.concat([feature,predict],axis=1)
```

```python
plt.scatter(r['first'],r['second'],r['third'],c=r['predict'], alpha=0.5)
centers = pd.DataFrame(model.cluster_centers_,columns=['first', 'second', 'third'])
center_x = centers['first']
center_y = centers['second']

plt.scatter(center_x,center_y,s=50, marker='D', c='r')
plt.show()
```
